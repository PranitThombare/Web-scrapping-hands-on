```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.thehindu.com/news/national/coronavirus-live-updates-may-29-2021/article34672944.ece?homepage=ture"
```

```python
page = requests.get(url)
```

```python
page
```

```python
soup = BeautifulSoup(page.content,'html.parser')
```

```python
print(page.text)
```

```python
img_tag =soup.find('source')
img_tag
```

```python
img_tag['srcset']
```

```python
img_url = img_tag['srcset']
```

```python
image = requests.get(img_url)
```

```python
with open('image.jpg','wb') as file:
    for chunk in image.iter_content(chunk_size=1024):
        file.write(chunk)
```
